### **Loading Images**

```python
# Load an image using 'imread' specifying the path to image
image1 = cv2.imread('image1.jpg')
```

### **Displaying Images**

```python
from matplotlib import pyplot as plt

#Show the image with matplotlib
plt.imshow(cv2.cvtColor(image1, cv2.COLOR_BGR2RGB))
plt.show()
```
